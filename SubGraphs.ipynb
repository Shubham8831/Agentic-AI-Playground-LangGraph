# **SubGraphs**

a subgraph is langgraph ususlly means a graph that is embedded and executed as node inside another graph

### Benifits
- modularity
- reusbility
- mentainability


### langgrpah benifits

- failure isolation( subgraph down then also execution complete BUT with warning) without subgraph failure can happen when node fails
- state seperation (all subgraph have there own state)
- observability

## Ways to create subgraph in LangGraph

### 1. invoke a graph from a node
- subgraphs are called from inside a node in the parent graph
- no direct connection both are build independently, invoke code in parent graph node
- have seperate state




### 2. add a grpah as a node
- a subgraph is added directly, added directly as a node in parent and share state keys with the parent
- have shared state
